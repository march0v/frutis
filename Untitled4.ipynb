```python
# ==============================
# 1. IMPORTAR LIBRERÍAS
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# ==============================
# 2. CARGAR ARCHIVO
# ==============================
from google.colab import files


# Cambia el nombre si es necesario
df = pd.read_csv("NKCollection_bd (1).csv")
```

```python
# ==============================
# 3. LIMPIEZA DE DATOS
# ==============================
print("Primeras filas:")
print(df.head())

print("\nInformación:")
print(df.info())

# Convertir variables categóricas a numéricas
df = pd.get_dummies(df, drop_first=True)
```

```python
# ==============================
# 4. DEFINIR VARIABLES
# ==============================
print(df.columns)  # Verifica el nombre correcto

# Separar características (X) y variable objetivo (y)
X = df.drop(columns=['Rendimiento'])
y = df['Rendimiento']

# Convertir X a numérico, coercing errores a NaN
X = X.apply(pd.to_numeric, errors='coerce')

# Combinar X e y temporalmente para eliminar filas con NaN de manera consistente
combined_df = pd.concat([X, y], axis=1)
combined_df = combined_df.dropna() # Eliminar filas con cualquier NaN en X o y

# Separar X e y de nuevo después de la limpieza
X = combined_df.drop(columns=['Rendimiento'])
y = combined_df['Rendimiento']
```

```python
# ==============================
# 5. DIVISIÓN DE DATOS
# ==============================
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.8, random_state=42
)
```

```python
# División
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.8, random_state=42
)

# Modelo
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X_train, y_train)

print("Modelo entrenado correctamente ")
```

```python
# ==============================
# 7. PREDICCIONES
# ==============================
y_pred = modelo.predict(X_test)
```

```python
# ==============================
# 8. MÉTRICAS
# ==============================
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n=== RESULTADOS ===")
print(f"Desviación promedio (MAE): {mae:.2f}")
print(f"Coeficiente R²: {r2:.4f}")
```

```python

# ==============================
# 9. COEFICIENTES
# ==============================
coeficientes = pd.DataFrame({
    'Variable': X.columns,
    'Coeficiente': modelo.coef_
}).sort_values(by='Coeficiente', ascending=False)

print("\n=== INFLUENCIA DE VARIABLES ===")
print(coeficientes)
```

```python

# ==============================
# 10. MATRIZ DE CORRELACIÓN
# ==============================
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title("Matriz de correlación")
plt.show()
```

```python

# ==============================
# 11. GRÁFICA REAL VS PREDICHO
# ==============================
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred)
plt.xlabel("Valores reales")
plt.ylabel("Predicciones")
plt.title("Real vs Predicho")
plt.show()
```

```python
# ==============================
# 12. INTERPRETACIÓN AUTOMÁTICA
# ==============================
print("\n=== INTERPRETACIÓN ===")

print(f"- El modelo tiene un error promedio de {mae:.2f} unidades.")
print(f"- Explica el {r2*100:.2f}% de la variabilidad del rendimiento.")

print("\nVariables más influyentes:")
print(coeficientes.head())

print("\nVariables menos influyentes:")
print(coeficientes.tail())
```
